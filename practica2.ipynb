```python

df['Descuento'] = df['precio'].apply(lambda precio: precio * 0.10 if precio > 500 else precio * 0.05)
df['Tipo'] = df['precio'].apply(lambda precio: 'Caro' if precio > 500 else 'Barato')
df
```

```python
#replace() sirve para reemplazar valores en una columna o DataFrame

#where sirve para filtrar datos basado en una condición, similar a un filtro booleano.
#usar where con drona sirve para mantener los valores que cumplen y eliminar los que no cumplen la condición que serian los NaN

#where en vez de salga nan se le puede poner un valor especifico o nombre

#df.dropna() elimina filas o columnas con valores nulos.
#df.dropna(subset=['precio']) elimina filas con valores nulos en la columna 'precio'.

#Rellenar valores nulos

#df.fillna('Desconocido) rellena valores nulos con 'Desconocido'
#df['precio'].fillna(df['precio'].mean()) rellena valores nulos en la columna 'precio' con la media de esa columna
```

```python
df.groupby('producto')['precio'].mean() #precio promedio por producto
```

```python
df.groupby('producto')['cantidad'].sum() #cantidad total vendida por producto
```

```python
df.groupby(['mes', 'producto'])['cantidad'].mean() #cantidad promedio vendida por producto cada mes
```

```python
df.groupby('producto').agg({
    'precio': ['mean', 'max', 'min'],
    'cantidad': 'sum'
})

#sirve para aplicar múltiples funciones de agregación a diferentes columnas en un DataFrame agrupado.
```

```python
df.sort_values('precio') #ordena por precio ascendente
df.sort_values('precio', ascending=False) #ordena por precio descendente

df.sort_values(['mes', 'precio'], ascending=[True, False]) #ordena por mes ascendente y precio descendente
```

```python
df.duplicated() #identifica filas duplicadas
df.duplicated().sum() #cuenta filas duplicadas

df.drop_duplicates() #elimina filas duplicadas
df.drop_duplicates(subset=['producto', 'mes']) #elimina filas duplicadas basadas en columnas específicas
```
