```python
import pandas as pd
```

```python
df = pd.read_excel('dataset-notas-estudiantes.xlsx')
```

```python
#Separamos los nombres ya que estaba unidos en un solo

df = df['Nombre,Matematicas,Ciencias,Espanol,Historia,Deportes'].str.split(',', expand=True)
```

```python
#se insertaron los nombres de las columnas
df.columns = ['Nombre', 'Matematicas', 'Ciencias', 'Espanol', 'Historia', 'Deportes']
```

```python
# se convirtio los nombres a minisculas ya que estaban en en diferentes formatos
df['Nombre'] = df['Nombre'].str.lower()

df
```

```python
cols = ['Matematicas', 'Ciencias', 'Espanol', 'Historia', 'Deportes']
```

```python
#Convertimos todos los datos a tipo numerico
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
```

```python
#Pasamos los valores negativos a valores absolutos
df[cols] = df[cols].abs()
```

```python
#Rellenamos los valores nulos con 0
df = df.fillna('0')
```

```python
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
```

```python
for col in cols:
    df[col] = df[col].apply(lambda x: x / 10 if x > 10 else x)
```

```python
df
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from scipy.spatial.distance import pdist
import matplotlib.pyplot as plt
```

```python
# Preparar datos numéricos para clustering (sin la columna 'Nombre')
import numpy as np
#Los valores
data = df[cols].values
```

```python
# Aplicar clustering aglomerativo usando el método de Ward
Z = linkage(data, method='ward')

plt.figure(figsize=(12, 6))
dendrogram(Z, labels=df['Nombre'].values, leaf_rotation=90)
plt.title('Dendrograma de Clustering Aglomerativo - Estudiantes', fontsize=14)
plt.xlabel('Estudiantes', fontsize=12)
plt.ylabel('Distancia', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Formar clusters
clusters = fcluster(Z, t=3, criterion='maxclust')
print("Cluster assignments:", clusters) 
```

```python
Z = linkage(data, method='ward')

distancias = []
k_values = range(2, len(data))

for k in k_values:
    clusters_temp = fcluster(Z, t=k, criterion='maxclust')
    suma_distancias = 0
    
    for i in np.unique(clusters_temp):
        puntos_cluster = data[clusters_temp == i]
        centroide = np.mean(puntos_cluster, axis=0)
        suma_distancias += np.sum(np.linalg.norm(puntos_cluster - centroide, axis=1)**2)
    
    distancias.append(suma_distancias)

plt.figure(figsize=(10, 6))
plt.plot(k_values, distancias, marker='o', linewidth=2, markersize=8, color='blue')
plt.xlabel("Número de clusters (k)", fontsize=12)
plt.ylabel("Suma de distancias al cuadrado (Inercia)", fontsize=12)
plt.title("Método del Codo para elegir k óptimo", fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

gradientes = np.gradient(distancias)
k_optimo = np.argmin(gradientes) + 2
plt.axvline(x=k_optimo, color='r', linestyle='--', linewidth=2, label=f'k óptimo sugerido = {k_optimo}')

plt.legend(fontsize=11)
plt.tight_layout()
plt.show()
```
